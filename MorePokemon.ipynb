```python
# min() and max() function
weakest_pokemon = pokemon.hp.min()
strongest_pokemon = pokemon.hp.max()
print(weakest_pokemon)
print(strongest_pokemon)
```

```python
# item() will give me the value inside the cell
# without the item() function, pandas returns an object
wimpy_pokemon = pokemon.loc[pokemon.hp == 1, 'name'].item()
wimpy_pokemon
```

```python
tough_pokemon = pokemon.loc[pokemon.hp == 255, 'name'].item()
tough_pokemon
```

```python
# getting a subsection of a dataframe - ie grabbing some rows
pokemon_army_idx = pokemon[[p for p in pokemon.hp > 150]]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(pokemon.hp, bins=25)
plt.ylabel("Number of Pokemon")
plt.title("Distribution of HP over all Pokemon")
plt.xlabel('HP')
```

```python
pokemon_groupings = pokemon.groupby('type1')
```

```python
water_pokemon = pokemon_groupings.get_group('water')
fire_pokemon = pokemon_groupings.get_group('fire')
electric_pokemon = pokemon_groupings.get_group('electric')
```

```python
type(water_pokemon)
```

```python
plt.hist(water_pokemon.hp, bins=5, label='Water')
plt.hist(fire_pokemon.hp, bins=5, label='Fire')
plt.xlabel("HP")
plt.ylabel("Number of Pokemon")
plt.legend()
```
