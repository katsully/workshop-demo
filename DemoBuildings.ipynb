```python
# tolist() will return a python list instead of a pandas object - lists are 
# easier to work with!
house_numbers = elevator_accidents_df['House Number'].tolist()
```

```python
# bad_places = [x + " " + y for x,y in zip(new_house_numbers, new_addresses)]

bad_places = [num + " " + street for num,street in zip(house_numbers, house_streets)]
```

```python
bad_places
```

```python
from collections import Counter
# get the five places that have the most elevator accidents 
worst_places = Counter(bad_places).most_common(5)
print(worst_places)
```
